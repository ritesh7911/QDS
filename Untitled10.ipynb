```python
import pandas as pd
import numpy as np
df = pd.read_excel(r'D:\Ritesh_new\retail_audit\sample_retail_data_ravi_for_model_2.xlsx')
# df['Store Name'].value_counts()
store_wise_df = []
df1 = df.copy()
for i in df1['Store Code'].unique():
    j = i
    df_j = df1[df1['Store Code']==j]
    x = df_j.reset_index(drop =True)
    store_wise_df.append(x)
prac = []
for dfs in store_wise_df:
    x = dfs.copy()
    x.set_index('Questions', inplace =True)
    y =x.copy()
    Yes_count =[]
    no_count = []
    weightage = []
    
    
    for i in dfs['Questions']:
        
        
        if i.find('Observation')!= -1:
            yes =0
            no =0
    #         print(y.loc[i][8])

            if y.loc[i][8]=='Yes':
                yes+=1
            else:
                no+=1
    #         print(y.loc[i][9])
            if y.loc[i][9]=='Yes':
                yes+=1
            else:
                no+=1
    #         print(y.loc[i][10])
            if y.loc[i][10]=='Yes':
                yes+=1
            else:
                no+=1
    #         print(yes)
    #         print(no)
            Yes_count.append(yes)
            no_count.append(no)
            weightage.append(np.nan)
            
        if i.find('Scored')!=-1:
            li = []
            max_ = 0
            min_ = 0

            li.append(y.loc[i][8])
            li.append(y.loc[i][9])
            li.append(y.loc[i][10])
            if li[0]>li[1]:
                if li[0]>li[2]:
                    max_ = li[0]
                else:
                    max_ = li[2]
            else:
                if li[1]>li[2]:
                    max_ = li[1]
                else :
                    max_ = li[2]
            Yes_count.append(np.nan)
        
            no_count.append(np.nan)
            weightage.append(max_)
#         print(Yes_count) 
#         print(no_count)
#         print(weightage)
    y['Number of yes'] = Yes_count
    y['Number of nos'] = no_count
    y['Weightage'] = weightage
    prac.append(y)
    
for i in prac:
    i.reset_index(inplace = True)
    
df_1 = pd.concat(prac)

df_1.reset_index(inplace =True)
df_1.drop(columns = 'index', inplace =True)

ob =[]

for i in df_1['Questions']:
    
    if i.find('Observation')!=-1:
        ob.append('pre')

    else:
        ob.append('abs')
        
df_1['Filter'] = ob
df11 = []
# print(len(df_1))
for i in df_1['Filter'].unique():
    j = i
    xx = df_1[df_1['Filter']==j]
#     print(type(xx))
    xx.reset_index(inplace =True)  
    df11.append(xx)
# df_2 = pd.concat(df11)
# df_2.reset_index(drop =True, inplace =True)
weightage = df11[1]['Weightage'].to_list()
df11[0]['Weights1'] = weightage

df_3 = df11[0].copy()
df_3.drop(columns = ['Weightage','Filter'],inplace = True)
df_3['Total responses'] = df_3['Number of nos']+df_3['Number of yes']
df_3.reset_index(drop =True, inplace =True)
df_3['Question_imp']=(df_3['Number of nos']/df_3['Total responses'])
df_3['Question_imp'] = df_3['Question_imp']*df_3['Weights1']
# df_3.group_by('Store Code')
df_3.drop(columns =['index'], inplace =True)

df_final = df_3.sort_values('Question_imp',ascending=False)
display(df_final)
```
