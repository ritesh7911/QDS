```python
import pandas as pd
import numpy as np
import re
import string

df_sfs = pd.read_excel(r'/home/saurabhtiwari/retail_data/VI/new_data/August_sfs.xlsx')
features = df_sfs.columns.to_list()
attributes = df_sfs.columns.to_list()
attributes_1 = attributes[:31]
attributes_2 = attributes[31:]
for i in attributes_2:
    if  i.find('Scored')== -1:
        pass
    else:
        attributes_1.append(i)
for i in attributes_2:
    if  i.find('Scorable')== -1:
        pass
    else:
        attributes_1.append(i)
        
df_sfs_new = df_sfs[attributes_1]
list_df =[]
for i in df_sfs_new.columns:
    if i.find('Scored')!=-1:
        list_df.append(pd.DataFrame(df_sfs_new[i]))
    else:
        pass
Question = []
weightage = []
no_count =[]
for i in list_df:
    x = i.columns.to_list()[0]
    Question.append(i.columns.to_list()[0].split(':')[0])
    t = str(i.max())
    
    u = t.split(':')[1].split('\n')[0]
    weightage.append(u)
    a = len(i[i[x]==0])
    b = len(i[x])
    no_count.append((a/b)*100)
df__ = pd.DataFrame({'Question':Question,'weightage':weightage,'Zero percentage':no_count})
df__['Multiplier'] = df__['weightage'].apply(lambda x:int(re.findall(r'\d+',x)[0]))

df__['IMP'] = df__['Zero percentage']*df__['Multiplier']
df__.sort_values('IMP',ascending=False)
```
